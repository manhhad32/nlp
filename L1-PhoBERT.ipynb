```python
import torch
from transformers import AutoModel, AutoTokenizer
from sklearn.metrics.pairwise import cosine_similarity

# Tải mô hình và tokenizer PhoBERT
model = AutoModel.from_pretrained("vinai/phobert-base")
tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base")

# Các câu ví dụ (đã được phân đoạn từ)
sentence1 = "Chúng_tôi là những nghiên_cứu_viên ."
sentence2 = "Chúng_tôi là các nhà khoa_học ."

# Mã hóa các câu
input_ids1 = torch.tensor([tokenizer.encode(sentence1)])
input_ids2 = torch.tensor([tokenizer.encode(sentence2)])

# Lấy biểu diễn ngữ nghĩa từ PhoBERT
with torch.no_grad():
    features1 = model(input_ids1)[0].mean(dim=1)
    features2 = model(input_ids2)[0].mean(dim=1)

# Tính toán độ tương đồng cosine
similarity = cosine_similarity(features1, features2)
print(f"Độ tương đồng giữa hai câu: {similarity[0][0]}")
```

```python
import torch
from transformers import AutoModel, AutoTokenizer

phobert = AutoModel.from_pretrained("vinai/phobert-base-v2")
tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base-v2")

# Tiền xử lý cho câu truy vấn và các văn bản
query = 'Chúng tôi đang học NLP'
documents = [
    'Chúng tôi đang học NLP ở Hutech',
    'Chúng tôi đang tìm hiểu PhoBERT',
    'Chúng tôi đang tìm hiều Doc2Vec',
    'Chúng tôi đang làm bài tập nhóm về NLP sử dụng mô hình PhoBERT và Doc2Vec'
]

# Biểu diễn câu truy vấn và các văn bản thành vector
query_input_ids = torch.tensor([tokenizer.encode(query)])
document_input_ids = [torch.tensor([tokenizer.encode(doc)]) for doc in documents]

with torch.no_grad():
    query_features = phobert(query_input_ids).last_hidden_state.mean(dim=1)
    document_features = [phobert(doc_ids).last_hidden_state.mean(dim=1) for doc_ids in document_input_ids]

# Tính độ tương đồng cosine
from sklearn.metrics.pairwise import cosine_similarity
similarities = [cosine_similarity(query_features, doc_features)[0][0] for doc_features in document_features]

# Kết luận: Văn bản có độ tương đồng cao nhất với câu truy vấn
most_similar_index = similarities.index(max(similarities))
most_similar_document = documents[most_similar_index]
print(f"Văn bản gần với câu truy vấn nhất: {most_similar_document}")
```
